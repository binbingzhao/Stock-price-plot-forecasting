# Stock Price Forecasting with Prophet

y(t) = g(t) + s(t) + h(t) + e(t)

- g(t): the growth function models non-periodic changes;
- s(t): periodic changes due to weekly or yearly seasonality;
- h(t): the effects of holidays
- e(t): the error term

```python
from fbprophet import Prophet
```

```python
AMZN
```

```python
# Dataframe must have"ds" and "y" with the dates and values (Prophet)

ph_amzn = AMZN.drop(['open', 'high','low','volume','Name'], axis=1)
ph_amzn.rename(columns={'close':'y','date':'ds'},inplace=True)

ph_amzn
```

# 📈 252 trading days a year

Technically there are 252 trading days out of a possible 365 in any given year in the U.S. stock market, which breaks down into about 4.85 trading days a week.


Total Trading Days per Year = 
365 - Weekends - Half Trading Days - Market Holidays

```python
train_size = 252*4 # Use 4 years trading data as train set
val_size = 252 # Use 1 year data as validation set
train_val_size = train_size + val_size
i = train_val_size # Day to forecast
```

```python
# fit prophet model
m = Prophet()
m.fit(ph_amzn[i-train_val_size:i])
```

```python
# Create dataframe with the forecasting days
future = m.make_future_dataframe(periods = 252) 
```

```python
# remove weekends for future df to improve the forecast
future['day'] = future['ds'].dt.weekday
future = future[future['day']<=4]
```

```python
predict = m.predict(future)
```

```python
# plot the forecast
m.plot(predict)
```

# Interactive EDA -- Plotly


```python
import plotly.offline as py
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)
from plotly import tools
import plotly.figure_factory as ff
```

```python
amzn = go.Scatter(x=AMZN.date, y=AMZN.close)
amzn = [amzn]
py.iplot(amzn)
```

```python
AAPL = sp500[sp500.Name=='AAPL']
FB = sp500[sp500.Name=='FB']

merge = AAPL[['date','close']].merge(FB[['date','close']], on = 'date')

merge['diff'] = merge['close_x'] - merge['close_y']
merge.head()
```

```python
trace0 = go.Scatter(x=AAPL.date, y=AAPL.close, name='AAPL')
trace1 = go.Scatter(x=FB.date, y=FB.close, name='FB')
trace2 = go.Scatter(x=merge.date, y=merge['diff'], name='DIFF')

data = [trace0, trace1,trace2]
py.iplot(data)
```

```python
trace0 = go.Box(
    y=AAPL.close,
    name = 'AAPL Close',
    marker = dict(
        color = 'grey')
)

trace1 = go.Box(
    y=FB.close,
    name = 'FB Close',
    marker = dict(
        color = 'royalblue')
)

data = [trace0, trace1]
layout = dict(title='Stock Prices of Apple & Facebook')

fig = dict(data=data, layout=layout)
py.iplot(fig)
```
